```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('../data/lighter_complete_deposit_txs.csv')
df.reset_index(drop=True)
pd.options.display.max_colwidth = 250
df.tail()
```

```python
dfWithdraws = pd.read_csv ('../data/lighter_complete_withdraw_txs.csv')
dfWithdraws.reset_index(drop=True)
```

```python
df = df[df["receipt_status"] == 1]
dfWithdraws = dfWithdraws[dfWithdraws["receipt_status"] == 1]
print(len(df),"successful deposits to TC contracts")
print(len(dfWithdraws),"successful withdraws from TC contracts")
print(len(df[df["transaction_type"]==2.0]),"deposits use EIP-1559 in our dataset.")
print(len(dfWithdraws[dfWithdraws["transaction_type"]==2.0]),"withdraws use EIP-1559 in our dataset.")
##Blocknative's gas fee estimation algorithm: max_fee = (2*base_fee) + max priority fee
walletFingerprintDeposits = df[df["max_fee_per_gas"] == 2*df["gas_price"]-df["max_priority_fee_per_gas"]]
walletFingerprintWithdraws = dfWithdraws[dfWithdraws["max_fee_per_gas"] == 2*dfWithdraws["gas_price"]-dfWithdraws["max_priority_fee_per_gas"]]
print(len(walletFingerprintDeposits), "deposits use Blocknative's gas fee estimation algorithm")
print(len(walletFingerprintWithdraws), "deposits use Blocknative's gas fee estimation algorithm")
```

```python
gwei = 1000000000
depositCounts = []
withdrawCounts = []
for i in range(0,10):
    walletFingerprintDeposits = df[df["max_priority_fee_per_gas"]==i*gwei]
    print(len(walletFingerprintDeposits), "deposits use custom set",i,"Gwei as max_priority_fees")
    walletFingerprintWithdraws = dfWithdraws[dfWithdraws["max_priority_fee_per_gas"]==i*gwei]
    print(len(walletFingerprintWithdraws), "withdraws use custom set",i,"Gwei as max_priority_fees")
    depositCounts.append(len(walletFingerprintDeposits))
    withdrawCounts.append(len(walletFingerprintWithdraws))
x = np.arange(0,10)
ax = plt.subplot(111)
ax.bar(x-0.1, depositCounts, width=0.2, color='b', align='center',label="Deposits")
ax.bar(x+0.1, withdrawCounts, width=0.2, color='r', align='center',label="Withdraws")
plt.legend()
plt.yscale("log")
plt.xlabel("Gwei")
plt.ylabel("Count")
plt.title("Transactions with round priority fees in Gwei")
plt.tight_layout()
plt.savefig("Figures/RoundPriorityFees.pdf")
plt.show()
```

```python
walletFingerprintDeposits = df.query("max_priority_fee_per_gas%1000000000!=0 & transaction_type==2.0")
print(len(walletFingerprintDeposits), "deposits use custom set max_priority_fees")
walletFingerprintWithdraws = dfWithdraws.query("max_priority_fee_per_gas%1000000000!=0 & transaction_type==2.0")
print(len(walletFingerprintWithdraws), "withdraws use custom set max_priority_fees")
```

```python
withdrawTxs = pd.read_csv ('../data/withdraw_transactions.csv')
withdrawTxs.reset_index(drop=True)
depositTxs = pd.read_csv('../data/deposit_transactions.csv')
depositTxs.reset_index(drop=True)
usedGasPercentageWithdraws = withdrawTxs ["receipt_gas_used"]/withdrawTxs ["gas"]
usedGasPercentageDeposits = depositTxs["receipt_gas_used"]/depositTxs["gas"]
differenceOfGasDeposits = depositTxs["gas"]-depositTxs["receipt_gas_used"]
differenceOfGasWithdraws = withdrawTxs ["gas"]-withdrawTxs ["receipt_gas_used"]
```

```python
ax = usedGasPercentageWithdraws.plot.hist(title="Used Gas Limit Histogram", alpha=0.5, legend=False, bins=200,color="red", label="Withdraws", logy=True, use_index=False)
ax = usedGasPercentageDeposits.plot.hist(title="Used Gas Limit Histogram", alpha=0.5, legend=False, bins=200, color="blue", label="Deposits", logy=True, use_index=False)
ax.legend()
plt.xlabel("Percentage of Used Gas")
ax.set_xlim(0,1)
plt.tight_layout()
plt.show()
```

```python
ax = differenceOfGasWithdraws.plot.hist(title="Difference of Gas Limit and Used Gas Histogram", alpha=0.5, legend=False, bins=200,color="red", label="Withdraws", logy=True, use_index=False)
ax = differenceOfGasDeposits.plot.hist(title="Difference of Gas Limit and Used Gas Histogram", alpha=0.5, legend=False, bins=200, color="blue", label="Deposits", logy=True, use_index=False)
ax.legend()
plt.xlabel("Difference")
plt.tight_layout()
plt.show()
```
